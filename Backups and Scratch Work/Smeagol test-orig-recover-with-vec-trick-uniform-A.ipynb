**Idea for lattice attack recovering $A$ from $s$ and $u$**

(Assuming $r$ and $e_1$ are correct)

Treating matrices by rows:

$$AS + E - T\equiv \mathbf{0} \pmod{q}$$
$$A^T\mathbf{r} + \mathbf{e_1} - \mathbf{u}\equiv \mathbf{0} \pmod{q}$$

equivalently:

$$AS + E - T\equiv \mathbf{0} \pmod{q}$$
$$\mathbf{r}A + \mathbf{e_1} - \mathbf{u}\equiv \mathbf{0} \pmod{q}$$

equivalently:

$$vec(A)samemult(S) + vec(E) - vec(T)\equiv \mathbf{0} \pmod{q}$$
$$vec(A)oppmult(\mathbf{r}) + vec(\mathbf{e_1}) - vec(\mathbf{u})\equiv \mathbf{0} \pmod{q}$$

or in block row form, $$(vec(A) | vec(E) | 1 | \mathbf{s'})\begin{pmatrix} samemult(S) \\ - \\ I \\ - \\ -vec(T) \\ - \\ qI \end{pmatrix} =  \mathbf{0} $$

and $$(vec(A) |  | 1 | \mathbf{s_1'})\begin{pmatrix} oppmult(r) \\ - \\ \\ - \\ vec(e_1)-vec(u) \\ - \\ qI \end{pmatrix} = \mathbf{0} $$

together:  $$(vec(A) | vec(E) |  |1 | \mathbf{s'} | \mathbf{s_1'})\begin{pmatrix} samemult(S) & oppmult(\mathbf{r})\\ - & - \\ I & 0\\ - & - \\  & \\ - & - \\ -vec(T) & vec(e_1)-vec(u)\\ - & - \\ qI & 0 \\ - & - \\ 0 & qI\end{pmatrix} =  \mathbf{0} $$
